```python
df=pd.read_excel('movie_data.xlsx')
```

```python
df
```

```python
data=df['上映时间'].value_counts()
```

```python
data
```

```python
x=data.index
y=data.values

plt.figure(figsize=(10,6))
plt.bar(x,y,color='g')
plt.title('不同年份电影数量',fontsize=20)
plt.xlabel('年份',fontsize=18)
plt.ylabel('电影数量',fontsize=18)
plt.tick_params(labelsize=14)
plt.show()#直方图
```

```python
data=df['上映时间'].value_counts()
data=data.sort_index()#排序
x=data.index
y=data.values
plt.figure(figsize=(10,6))
plt.plot(x,y,color='b')
plt.title('每年电影数量',fontsize=20)
plt.ylabel('数量',fontsize=18)
plt.xlabel('年份',fontsize=18)
plt.tick_params(labelsize=14)

for a,b in zip(x[::4],y[::4]):
    plt.text(a,b,b,ha='center',va='bottom',fontsize=13)#每隔三年在曲线上标注数量
plt.show()#折线图
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['评分'],bins=15,edgecolor='k')#十五个柱子
plt.show()#直方图
```

```python
import matplotlib.mlab as mlab
from scipy.stats import norm
```

```python
fig=plt.figure(figsize=(10,8))
ax1=fig.add_subplot(111)
n,bins,patches=ax1.hist(df['评分'],bins=15,color='m')
ax1.set_ylabel('电影数量',fontsize=15)
ax1.set_xlabel('评分',fontsize=15)
ax1.set_title('频率分布图',fontsize=20)


y=norm.pdf(bins,df['评分'].mean(),df['评分'].std())#概率密度
ax2=ax1.twinx()
ax2.plot(bins,y,'b--')
ax2.set_ylabel('概率密度',fontsize=15)
plt.show()
```
